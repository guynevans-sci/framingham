# 3.5  Heating
I love this section: The result is that blueberry earth will turn into a roaring ocean of boiling jam, with geysers of released air and steam likely ejecting at least a few berries into orbit. But clearly water can’t heat up to 143.5793 K...or can it?  And what’s all this about supercritical water?  We need a phase diagram, Figure 2.  Let’s make the simple assumptions that:
* heating of BBE during coalescence is either:  
  - (a) **isothermal** (dp/dT = 1/0 GPa K<sup>−1</sup>) OR
  - (b) **adiabatic** (dp/dT= 1/20 GPa K<sup>−1</sup>)

* Regardless the surface temperatures rise approximately 140 K above the pre-coalseced T<sub>surf</sub>=290 K.
* The pure-water phase diagram is close enough for use here.

Plot the likely geotherm for BBE from P= 0 to P= 4.5 GPa. Does, as Sandberg hypothesize, BBE get a frozen, sweet, blueberry-water mixture (a granita) for a core?  Under what circumstances? At what depth(s)?

```python

surface_temperature = 290+140
geotherm_slope = 1/20
```

```python
#Import the appropriate libraries
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
#Define file location of water phase lines data.
PL_fileloc  = 'WaterPhaseLines.csv'
```

```python
#Read this data into pandas (a python library for manipulating data tables)
phase_lines = pd.read_csv(PL_fileloc, low_memory=False)
```

```python
#Setting up the figure
fig, ax = plt.subplots()
ax.set_xlabel("Temperature (K)")
ax.set_ylabel("Pressure (GPa)")
plt.title('The H2O phase diagram')
plt.xlim(200, 600)   
plt.ylim(0, 4.5)
plt.text(400,1.25,'Liquid')
plt.text(230,0.05,'Ih')
plt.text(215,0.35,'II')
plt.text(245,0.25,'III')
plt.text(250,0.43,'V')
plt.text(230,1.25,'Ice VI')
plt.text(300,3.25,'Ice VII')
ax.figure.set_size_inches(15, 10)
font = {'size'   : 18}
plt.rc('font', **font)

#Plotting the water phase diagram
number_of_intervals = 10
for row in range(0, phase_lines.shape[0]):
    T_min_K = phase_lines.values[row, 1]
    T_max_K = phase_lines.values[row, 2]
    beta_0 =  phase_lines.values[row, 3]
    beta_1 =  phase_lines.values[row, 4]
    beta_2 =  phase_lines.values[row, 5]
    T = list([T_min_K + (T_max_K - T_min_K)/number_of_intervals*i for i in range(0,number_of_intervals+1)])
    P = list([beta_0 + beta_1*i + beta_2*i**2 for i in T])
    ax.plot(T, P, 'b--')

#Plotting the geotherm
geotherm_pressure = [i/2 for i in range(0,10)]
geotherm_temperature = [(pressure / geotherm_slope) + surface_temperature for pressure in geotherm_pressure]
ax.plot(geotherm_temperature, geotherm_pressure, 'r-')
geotherm_rotation = math.degrees(math.atan(geotherm_slope * (10/15) / (4.5/400)))
plt.text(2.25 / geotherm_slope + surface_temperature - 25,2.25,'Geotherm', color = 'red', rotation = geotherm_rotation)
                              
plt.show()
```
